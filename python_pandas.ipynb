# Pandas

```python
# Pandas ::=>
# - Used for data analysis and data manipulation
# - Here we have two categories
#   1. Series -> Nothing but column
#   2. DataFrame -> Collections of series
```

# 1.Series

```python
# Series =>
# - A Pandas Series is a one-dimensional labeled array
# Creating Series ->
# ------------------
# 1.creating a series from list
# Ex=>

import pandas as pd

x = [10,20,30,40,50]
series = pd.Series(x)
print(series)
```

```python
# 2.creating a series from dictionary
# Ex=>

import pandas as pd

data = {"maths":90,"pyh":89,"chem":85}
series = pd.Series(data)
print(series)
```

```python
# 3.creating our own index
import pandas as pd

series = pd.Series([10,20,30,40,50],index=["a","b","c","d","e"])
print(series)
```

# 2.DataFrame

```python
# DataFrame ::=> Collection of series is called DataFrame
# Creating DataFrame -
#
# 1. creating DataFrame from dictionary
# Ex=>

import pandas as pd

data = {
    "List-1":[90,80,70,60,50,40],
    "List-2":[60,50,40,30,20,10]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Ex=>

import pandas as pd

data = {
    "Ename":["Ankit","Chintu","Kalyan","Lali"],
    "Salary":[200000,100000,300000,250000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# our own index
import pandas as pd

data = {
    "Sony":[250,300,200,250,350,400],
    "LG":[300,180,260,290,600,250],
    "Lenvo":[450,320,420,190,250,150]
}

df = pd.DataFrame(data,index=["Jan","Feb","Mar","Apr","May","Jun"])
print(df)
```

```python
# Retrieving praticular month data(particular row) using ioc
# Ex=>
print(df.loc['Apr'])
```

```python
# Indexing and Slicing
#
# 1. Printing a specific column(series)
# Ex=>
print(df['LG'])

# 2. Printing a DataFrame
# Ex=>
print(df[['LG']])

# 3. Printing multiple column as DataFrame
print(df[['LG',"Sony"]])

# 4. Slicing a DataFrame
x = df[0:3] # result is alsp DF,here 3 is upper boung excluded
print(x)
```

```python
# More data selection operations using -
# 1. loc -> uses string index
# 2. iloc -> uses integer index
# Ex ==> loc

x = df.loc[['Apr','May']]
print(x)

# Ex==>iloc
x = df.iloc[3]  # fetch the 4th row
print(x)
```

```python
# getting more than one column
# Ex=>
x = df.iloc[:]  # fetch all the rows
print(x)

x = df.iloc[:2] # two row all column
print(x)

x = df.iloc[:,1:3] # all rows $ 1-2 columns
print(x)

x = df.iloc[0:3,1:3] # 0-2 rows & 1-2 columns
print(x)

x = df.iloc[0:3,0:3:2] # 0-2 rows,0-2 columns increment by 2
print(x)

# To extract 250 from 1 rows
x = df.iloc[0:1,0:1]
print(x)
```

```python
# Sorting Data
# sort_values() ::
# => By default sorting irder is asceding order
# ex=>

import pandas as pd

data = {
    "Sony":[250,300,200,250,350,400],
    "LG":[300,180,260,290,600,250],
    "Lenvo":[450,320,420,190,250,150]
}

df = pd.DataFrame(data,index=["Jan","Feb","Mar","Apr","May","Jun"])
print(df)

# sort by Sony
x = df.sort_values('Sony')
print(x)
```

```python
# DF filtering operations

import pandas as pd

data = {
    'Name':['Ankit','Chintu','Lali','Tanu','Kalyan','Poece'],
    'Age':[21,22,21,24,23,22],
    'Gender':['Male','Male','Female','Female','Male','Female'],
    'Branch':['CSE','EEE','EEE','IT','CSE','IT'],
    'Percentage':[84,86,75,89,90,95]
}

# create DF
df = pd.DataFrame(data,
                  columns=['Name','Age','Gender','Branch','Percentage'],
                  index=['s1','s2','s3','s4','s5','s6'])
print(df)
```

```python
# 1. Filtering data based on percentage

df1 = df[df['Percentage']>85]
print(df1)
print('\n')

# if we filter a DF then result also a DF

# (or)

df1 = df.loc[df['Percentage']>85]
print(df1)
print('\n')
```

```python
# 2. I want only the CSE students

df2 = df[df['Branch']=='CSE']
print(df2)
```

```python
# 3. I want other than CSE

df3 = df[df['Branch']!='CSE']
print(df3)
```

```python
# 4.isin() method :: to check whether each element in a Series or
#                    DataFrame is present in a given list
# Ex => Require only those student who belongs to CSE or IT

options = ['CSE','IT']
df1 = df[df['Branch'].isin(options)]
print(df1)
print('\n')

# (or)

df1 = df.loc[df['Branch'].isin(options)]
print(df1)
```

```python
# 5. I want who are from non-CSE

df1 = df[~df['Branch'].isin(options)]
print(df1)
```

```python
# Providing multiple conditions
# Ex=> whose % <70 and non-CSE
options=['CSE']
df1 = df[(df['Percentage']<90) & (~df['Branch'].isin(options))]
print(df1)
```

```python
# Grouping and Aggregation
import pandas as pd

df = pd.read_csv('/content/orders.csv')
print(df)
```

```python
# Reading specific row or a particular columns -> 5 row
print(df[0:5]['customer'])

# Reading customer or order for all rowa using loc
#print(df.loc[:,['customer','order']])

# Reading customer or order_state for 11 rows using loc
print(df.loc[0:10,['customer','order_state']])

```

```python
# Reading specific rows and columns
print(df.loc[[1,4,6],['customer','order','order_state']])

# Specific columns for a range of rows
print(df.loc[2:6,['customer','order','order_state']])
```

```python
# Various Aggregatio-> sum,mean,max,min,count
# 1. Number of order -> using count()
order_count = df['order'].count()
print("Number of order : \n",order_count)

# 2. Total order value -> using sum()
total_order_value = df['revenue'].sum()
print("Total order value : \n",total_order_value)

# 3. Average order value -> using mean()
avg_order_value = df['revenue'].mean()
print("Avf order value : \n",avg_order_value)

# 4. Max order value -> using max()
max_order_value = df['revenue'].max()
print("Max order value : \n",max_order_value)

# 5. Min order value -> using min()
min_order_value = df['revenue'].min()
print("Min order value : \n",min_order_value)
```

```python
# Grouping
# Task-1 : Calculate total number of order per customer
order_count = df.groupby('customer')['order'].count()
#print("Total number of order/customer : \n",order_count)

# Task-2 : Calculate total number of revenue per customer
tov_cust = df.groupby('customer')['revenue'].sum()
#print("Total number of revenue/customer : \n",tov_cust)

# Task-3 : Calculate count of order based onf order_state
ord_count = df.groupby('order_state')['order'].count()
print('order count : \n',ord_count)

# Task-4 : AOV per customer
aov_cust = df.groupby('customer')['revenue'].mean()
#print('AOV per customer : \n',aov_cust)
```
