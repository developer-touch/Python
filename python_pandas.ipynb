# Pandas

```python
# Pandas ::=>
# - Used for data analysis and data manipulation
# - Here we have two categories
#   1. Series -> Nothing but column
#   2. DataFrame -> Collections of series
```

# 1.Series

```python
# Series =>
# - A Pandas Series is a one-dimensional labeled array
# Creating Series ->
# ------------------
# 1.creating a series from list
# Ex=>

import pandas as pd

x = [10,20,30,40,50]
series = pd.Series(x)
print(series)
```

```python
# 2.creating a series from dictionary
# Ex=>

import pandas as pd

data = {"maths":90,"pyh":89,"chem":85}
series = pd.Series(data)
print(series)
```

```python
# 3.creating our own index
import pandas as pd

series = pd.Series([10,20,30,40,50],index=["a","b","c","d","e"])
print(series)
```

# 2.DataFrame

```python
# DataFrame ::=> Collection of series is called DataFrame
# Creating DataFrame -
#
# 1. creating DataFrame from dictionary
# Ex=>

import pandas as pd

data = {
    "List-1":[90,80,70,60,50,40],
    "List-2":[60,50,40,30,20,10]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Ex=>

import pandas as pd

data = {
    "Ename":["Ankit","Chintu","Kalyan","Lali"],
    "Salary":[200000,100000,300000,250000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# our own index
import pandas as pd

data = {
    "Sony":[250,300,200,250,350,400],
    "LG":[300,180,260,290,600,250],
    "Lenvo":[450,320,420,190,250,150]
}

df = pd.DataFrame(data,index=["Jan","Feb","Mar","Apr","May","Jun"])
print(df)
```

```python
# Retrieving praticular month data(particular row) using ioc
# Ex=>
print(df.loc['Apr'])
```

```python
# Indexing and Slicing
#
# 1. Printing a specific column(series)
# Ex=>
print(df['LG'])

# 2. Printing a DataFrame
# Ex=>
print(df[['LG']])

# 3. Printing multiple column as DataFrame
print(df[['LG',"Sony"]])

# 4. Slicing a DataFrame
x = df[0:3] # result is alsp DF,here 3 is upper boung excluded
print(x)
```

```python
# More data selection operations using -
# 1. loc -> uses string index
# 2. iloc -> uses integer index
# Ex ==> loc

x = df.loc[['Apr','May']]
print(x)

# Ex==>iloc
x = df.iloc[3]  # fetch the 4th row
print(x)
```

```python
# getting more than one column
# Ex=>
x = df.iloc[:]  # fetch all the rows
print(x)

x = df.iloc[:2] # two row all column
print(x)

x = df.iloc[:,1:3] # all rows $ 1-2 columns
print(x)

x = df.iloc[0:3,1:3] # 0-2 rows & 1-2 columns
print(x)

x = df.iloc[0:3,0:3:2] # 0-2 rows,0-2 columns increment by 2
print(x)

# To extract 250 from 1 rows
x = df.iloc[0:1,0:1]
print(x)
```

```python
# Sorting Data
# sort_values() ::
# => By default sorting irder is asceding order
# ex=>

import pandas as pd

data = {
    "Sony":[250,300,200,250,350,400],
    "LG":[300,180,260,290,600,250],
    "Lenvo":[450,320,420,190,250,150]
}

df = pd.DataFrame(data,index=["Jan","Feb","Mar","Apr","May","Jun"])
print(df)

# sort by Sony
x = df.sort_values('Sony')
print(x)
```

```python
# Renaming a DataFrame column
# Ex=> Sony -> Samsung

x = df.columns=['Samsun','LG','Lenvo']
print(x)

# Ex=> Lenvo -> MAC
x = df.columns=['Sumsung','LG','MAC']
print(x)
```

```python
# Dropping a column
x = df.drop('LG',axis=1)
print(x)
print(df)
```

```python
# Dropping a row
x = df.drop('May')
print(x)
```

```python
# Adding new column
# Ex=>

df['vivo'] = df.eval('LG+150')
print(df)

# Adding new column - Mi
df['Mi'] = pd.Series([350,240,340,280,130,450],
              index=['Jan',"Feb","Mar",'Apr','May','Jun'])
print(df)
```

```python
# Converting DF to another format
df.to_excel('mobile.xlsx')
df.to_csv('mobile.csv')
df.to_json('mobile.json')

# find Max
x = df.LG.max()
print(x)

# find missing values
# isnull() -> will tell if a column misses a value
# Ex=>
print(df.isnull())

# find rank
# rank() :: to rank every column according to its value
# Ex=>
print(df.rank())
```

```python
# concatenatiing
# concat() :: we can concatenate 2 or more DF
x = pd.concat([df,df])
print(x)

# Note :: we can concatenate 3 or more DF
```
