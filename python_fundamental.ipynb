# 8.Tuple

```python
# Tuple :: Tuple is a collection datatype which hold group of value
# Properties ::
# 1 -> Tuple is represented as ()
# 2 -> Tuple supported heterogeneous elements and also support duplicate
# 3 -> Tuple follow the insertion order and supports 2 type of index
#    1. Positive Index, index start from 0
#    2. Negative Index, index start from -1
# Ex =>

x = (10,"Ankit",21)  # heterogeneous
y = (10,20,30,10,20) # duplicates

# 3 -> Tuple is immutable object

x = (10,20,30,40,50)
print(x,type(x))

# x[1] = 15 -> # we can't change Tuple object

x = 10
print(x,type(x))

x = 10,20,30,40   # without () by default its take Tuple
print(x,type(x))

cities = "Chennai","Hyd","Patna","Delhi"
print(cities,type(cities))

a = ("Python") # if only one element in a tuple the its not a tuple
               # whichever element is present it take that type
print(a,type(a))

b = ("Python",) #Tuple
print(b,type(b))

c = (10)  # int
print(c,type(c))

d = 10, #Tuple
print(d,type(d))

e = "Python"
print(e,type(e),len(e))

f = "Python",
print(f,type(f),len(f))

g = ([10,20,30])
print(g,type(g),len(g))

h = ([10,20,30],)
print(h,type(h),len(h))

i = "hello"
print(i*3)

j = (10,20,30)
k = (40,50,60)
l = j+k
print(l,type(l),len(l))
```

# 9.Set

```python
# Set ::=> A Set is a collection of datatypes which holds group of value
# Properties =>
# 1 -> A set is represented by {}
# 2 -> Set allow heterogeneous elements
# 3 -> Set doesn't allows duplicates
# 4 -> In Set,Insertion order is not followed
# 5 -> Set doesn't support indexin
# 6 -> Set are mutable object i.e changes or modification are allowed
#      within a set but the element of set should be always immutable
# Ex =>

# x = {10,20,30.5,[1,2,3]} here list not allowed bcox mutable

x = {10,20,30,40,50}
print(x,type(x))

# y = {10,20,30,40,10} duplicates are not allowed

y = {10,30.5,(10,20,30)}
print(y,type(y),len(y))
```

```python
# Set Function

x = {50,30,10,20,40}
print(x,type(x))
print(x,len(x))
print(x,max(x))
print(x,min(x))
print(x,sum(x))
print(sorted(x))
print(sorted(x,reverse=True))
```

```python
# Set Methods

x = {10,20,30,40,50}
print(x,len(x))

# 1. add() ::=> To add an element to a set
x.add(60)
print(x,len(x))

x.add((5,15,25))
print(x,len(x))

# 2. update() ::=> For adding multiple element at a time
x.update((35,45,55))
print(x,len(x))

# 3. copy() ::=> For duplicating a set
y = x.copy()
print(y,len(y))

# 4. pop() ::=> Removes an element randomly
print(x.pop())
print(x,len(x))

print(y.pop())
print(y,len(y))

# 5. remove() ::=> For removing a particular element from set
#             ::=> if not present raise error
x.remove(20)
print(x,len(x))

# 6. discard() ::=> For removing a particular element from set
#             ::=> if not present doesn't raise error
x.discard(30)
print(x,len(x))

#7. clear() :: Cleara all the set elements
x.clear()
print(x,len(x))
```

```python
# Set Mathematical Methods

x = {10,20,30,40,50}
y = {30,40,50,60,70}

# 1. union()
z = x.union(y)
print(z,len(z))

# 2. intersection()
z = x.intersection(y)
print(z,len(z))

# 3. difference()
z = x.difference(y)
print(z,len(z))

# or
z = x-y
print(z,len(z))

# 4. symmetric_difference :: excluding the common elements
z = x.symmetric_difference(y)
print(z,len(z))
```

# 10.Dictionary

```python
# Dictionary :: Dictionary is collection of (k,v) pairs
# Properties ::
# 1 -> Dictionary represented as {}.
x = {"Name":"Ankit","Age":21}
# 2 -> In dictionary keys can't be duplicated,values can be
# 3 -> Dictionary doesn't support indexing
# 4 -> Dictionary is un-oreder and mutable
# Ex =>

x = {"Name":"Ankit","Age":21,"City":"Patna"}
print(x,type(x))

# if we pass key -> it returns the value
print("Name = ",x["Name"])
print("Age = ",x["Age"])
print("City = ",x["City"])
```

```python
# Modifying,Adding,Printing values of a dictionary
x = {"Name":"Ankit","Age":21,"City":"Patna","Wt":53}
print(x,len(x))

# modifying
x["Wt"] = 52
print(x,len(x))

# adding a new element -> designation
x["designation"] = "Manager"
print(x,len(x))

# adding one more elements -> marks
x["Marks"] = [["Maths",90],["Phys",88],["Chem",89]]
print(x,len(x))

# modifying the marks[maths] 90 -> 95
x["Marks"][0][1] = 95
print(x,len(x))

# modifying the marks[phys] 88 -> 90
x["Marks"][1][1] = 90
print(x,len(x))

# dictionary within a dictionary
y = {"Name":"Ankit","Age":21,
     "Children":{"Ch1":1,"Ch2":2},
     "Address":{"City":"Patna","State":"Bihar","Pin":844102}
    }
print(y,len(y))

# print Patna
print(y["Address"]["City"])

# add country to address
y["Address"]["Country"]= "IND"
print(y,len(y))
```

```python
# Dictionary Methods
x = {"Name":"Ankit","Age":21,"City":"Patna","Wt":53}
print(x,len(x))

# 1. keys() ::=> prints all the keys
print(x.keys())

# 2. values() ::=> prints all the values
print(x.values())

# 3. pop() ::=> removinng an element from dictionary
print(x.pop("Wt"))
print(x,len(x))

# 4. popitem() ::=> removes an elements randomly
print(x.popitem())
print(x,len(x))

# 5. clear() ::=> clears all the element of a dictionary
x.clear()
print(x,len(x))
```

```python
# Advantage of return statements
# => The computed result of a function,if to be used in the later
#    part of the program
#
# Ex => Compute tax(10%) and also copute net salary

salary = 8000
def compute():
  tax = salary * 0.10
  return tax

result = compute()
net_salary = salary - result
print("Net Salary ",net_salary)
```

```python
# Can function return 2 values -> YES

def compute(s1,s2,s3):
  total = s1+s2+s3
  avg = total/3
  return total,avg

result = compute(90,85,95)
print(result,type(result))
print("-----------------")
print("Total = ",result[0])
print("Avg = ",result[1])
print("---------------")
#or
total,avg = compute(90,85,95)
print("Total = ",total)
print("Avg = ",avg)
```
