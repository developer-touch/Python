# Numpy

```python
# Numpy =>
# -> Numpy is stands for Numerical Python
# -> Numpy is to perform mathematical and statistcal operations
# -> It has been built to work with N-dimensional array,linear
#    algebra,random number and etc...
# Ex=>
import numpy as np

# To know numpy version
print(np.__version__)

# => Numpy arrays are faster as compared with lists especially
#    when working with larger datasets
```

# 1.1D Array

```python
# 1.Creating a Numpy Array from a list

list = [10,20,30,40,50]
print(list,type(list))

array1 = np.array(list)
print(array1,type(array1))

# or
array1 = np.array([5,15,25,35,45])
print(array1,type(array1))
```

```python
# 2. Performing methematical operations like sum,sub,mul,div
# add
add = array1+10
print(add)

# sub
sub = add-1
print(sub)

# mul
mul = sub*2
print(mul)
```

```python
# 3. Shape of the array and dtype
print(array1.shape) # return no of elements
print(array1.dtype)
```

```python
# 4. float type
array2 = np.array([1.5,2.5,3.5,4.5,5.5])
print(array2)
print(array2.shape)
print(array2.dtype)
```

# 2.2D Array

```python
# 5. 2 dimensional array
array3 = np.array([
    (2,4),
    (3,5)
])
print(array3)
print(array3.shape)
print(array3.dtype)
```

# 3D Array

```python
# 6. 3 dimensional array
array4 = np.array([
    [
      (1,2,3),
      (3,4,5)
    ],
    [
      (5,6,7),
      (7,8,8)
    ],
    [
      (10,11,12),
      (13,14,15)
    ]
])
print(array4)
print(array4.shape)
print(array4.dtype)
```

# 4.Zeros & Ones Array

```python
# 7. np.zeros & np.ones
# -> we can create a matric full of ones and zeros using
#    1. np.zeros -> for zeros
#    2. np.ones  -> for ones
# -> it can be used when you initialized the weight during
#    first iteration in Transflow & other statistical tasks.
# Syntax=>
#   np.zeros(shape,dtype)
import numpy as np

array1 = np.zeros((2,2))
print(array1)

print(np.zeros((2,2),dtype=int))
print(np.ones((2,2,3),dtype=int))
```

# 5.Reshape Array

```python
# reshape() ->In some occasiona,need to reshape the data from wide to long
# Syntax =>
#    np.reshape(array,newShape)
# Ex=> pythonDB,ML analysis

array1 = np.array([
    (2,3,4),
    (3,4,5)
])
print(array1)
print(array1.shape)

# reshape the array1
print(array1.reshape(3,2))
#print(array1.reshape(4,2))
print(array1.reshape(6,1))
#print(array1.reshape(4,1))
```

# 6.Flatten Array

```python
# flatten() -> Flattening the array, when you deal with some neural
#              network like,we need to flatten the array
# Ex=>
print(array1)
print(array1.flatten())
```
